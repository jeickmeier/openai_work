```python
import asyncio
from crawl4ai import AsyncWebCrawler
from urllib.parse import urljoin, urlparse

async def crawl(url, depth, visited):
    if depth == 0 or url in visited:
        return
    visited.add(url)
    async with AsyncWebCrawler(verbose=True) as crawler:
        result = await crawler.arun(url=url)
        if result.success:
            print(f"Crawled: {url}")
            for link in result.links['internal']:
                full_url = urljoin(url, link['href'])
                # Ensure the link is within the same domain
                if urlparse(full_url).netloc == urlparse(url).netloc:
                    await crawl(full_url, depth - 1, visited)
        else:
            print(f"Failed to crawl: {url}")
    
    return result
```

```python
start_url = "https://www.antares.com/market-insights/"
max_depth = 2
visited = set()
result = await crawl(start_url, max_depth, visited)
Markdown(result.markdown)
```
